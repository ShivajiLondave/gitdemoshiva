```
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Bob', None],
    'Age': [25, None, 35, 30, 22],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Mumbai', 'Pune']
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

df = df.drop_duplicates()

df = df.dropna(subset=['Name'])

df['Age'] = df['Age'].fillna(df['Age'].mean())

print("\nCleaned Data:")
print(df)
```
